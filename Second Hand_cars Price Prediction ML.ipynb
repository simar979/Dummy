```python
#z score----->
#detect---->in km driven---->find z score(outliers)
#calculate z score for each data point in the column
#z score how many standard deviations a data point is away from the the mean
```

```python
def out_detect(column): #out_detect function applied to identify outliers
    mean=column.mean()
    std=column.std()
    z_score=(column-mean)/std
    return column[(z_score>3) | (z_score<-3)]

outlier=out_detect(numeric["km_driven"])
f=numeric["km_driven"].isin(outlier) #f is created to identify rows and columns
numeric=numeric[~f] # numeric and category are filtered to remove rows with the outliers 
category=category[~f]

outlier1=out_detect(numeric["max_power"])
f=numeric["max_power"].isin(outlier1)
numeric=numeric[~f]
category=category[~f]

outlier2=out_detect(numeric["mileage"])
f=numeric["mileage"].isin(outlier2)
numeric=numeric[~f]
category=category[~f]
```
